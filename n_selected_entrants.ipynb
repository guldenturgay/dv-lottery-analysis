```python
import pandas as pd
```

```python
DV_selected_entrants_16 = pd.read_csv('DV_selected_entrants_2016.csv', header = None)
DV_selected_entrants_17 = pd.read_csv('DV_selected_entrants_2017.csv', header = None)
DV_selected_entrants_18 = pd.read_csv('DV_selected_entrants_2018.csv', header = None)
```

```python
DV_selected_entrants_16.columns = ['Country', '2016']
DV_selected_entrants_17.columns = ['Country', '2017']
DV_selected_entrants_18.columns = ['Country', '2018']
```

```python
# Separate numbers from country names and create another column for numbers

def clean_table(df):
    df.iloc[:, 1] = df.iloc[:, 0].str.extract(r'(\d.*)')
    df.iloc[:, 0] = df.iloc[:, 0].str.extract(r'(\D*\s)')
    # after using encode('utf-8') it appears that the Country names are not consistent on the tables
    df.iloc[:, 0] = df.iloc[:, 0].str.replace('\xa0', ' ')
    return df
    
```

```python
clean_table(DV_selected_entrants_16)
clean_table(DV_selected_entrants_17)
clean_table(DV_selected_entrants_18)
```

```python
# Merge dataframes 

pd.set_option('display.max_rows', 200)
DV_16_17 = DV_selected_entrants_16.merge(DV_selected_entrants_17, how = 'left', on = 'Country')
DV_selected_entrants = DV_16_17.merge(DV_selected_entrants_18, how = 'left', on = 'Country')

# Fill NaN values with 0

DV_selected_entrants = DV_selected_entrants.fillna(0)
DV_selected_entrants
```
